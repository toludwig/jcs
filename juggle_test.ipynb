```python
%load_ext autoreload
%autoreload 2
from juggler import *
from model import SAC
import matplotlib.pyplot as plt
```

```python
PATTERN = [3,3,3]

N_STEP = 1000

env = Juggler(PATTERN, rendering=False, verbose=False)
agent = OptimalAgent(PATTERN)
#agent = SAC(Juggler.state_dim, Juggler.action_dim) # PATTERN TODO

drop = False
rewards = []
step = 0
obs = env.reset()
while not drop and step < N_STEP:
    ctrl = agent.act(obs)
    obs, reward, drop = env.step(ctrl)
    rewards += [reward]
    step += 1
```

```python
plt.plot(rewards)
```
