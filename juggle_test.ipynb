```python
%load_ext autoreload
%autoreload 2
import numpy as np
import matplotlib.pyplot as plt
from juggler import Juggler
```

```python
DT = 0.01
V0 = 5
GRAVITY = -10
THETA = np.radians(90) # throw angle
# ALPHA = 0.05 # rho * A / 2m
# C_LIFT = -0.1
# C_DRAG = 0

def trajectory(x, v): # position, velocity, acceleration
    # v[0] += DT * -ALPHA * (C_DRAG * v[0] + C_LIFT * v[1]) * np.sqrt(v[0]**2 + v[1]**2)
    # v[1] += DT * (ALPHA * (-C_DRAG * v[1] + C_LIFT * v[0]) * np.sqrt(v[0]**2 + v[1]**2) - GRAVITY)
    v += DT * np.array([0, GRAVITY])
    x += DT * v
    return (x, v)

x0 = np.zeros(2)
v0 = np.array([np.cos(THETA), np.sin(THETA)]) * V0

xs = [x0]
x, v = trajectory(x0, v0)
while x[1] >= 0:
    xs += [x.copy()]
    x, v = trajectory(x, v)

xs = np.vstack(xs)
# print(xs)
plt.figure()
plt.scatter(xs[:,0], xs[:,1])
plt.show()
```
