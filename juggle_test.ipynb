```python
%load_ext autoreload
%autoreload 2
from juggler import *
import matplotlib.pyplot as plt
```

```python
PATTERN = [3,3,3]
N_STEP = 350

env = Juggler(PATTERN, rendering=False, verbose=False)
agent = OptimalAgent(PATTERN)

ctrl = agent.control(env.get_state())
drop = False
rewards = []
step = 0
while not drop and step < N_STEP:
    state, drop, reward = env.step(*ctrl)
    ctrl = agent.control(state)
    rewards += [reward]
    step += 1
```

```python
plt.plot(rewards)
```
